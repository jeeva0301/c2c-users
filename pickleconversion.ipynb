```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Load your data
df = pd.read_csv("c2c.csv")
df=df[['socialNbFollows', 'productsListed', 'productsSold', 'productsPassRate', 'productsWished', 'productsBought', 'civilityGenderId', 'hasAnyApp', 'daysSinceLastLogin', 'seniority']]
from sklearn.preprocessing import MinMaxScaler,OrdinalEncoder

# Min-Max Scaling
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
print(X.columns)
import pickle
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Assume that your data is stored in a variable called "X_new"

# Create the pipeline
pipeline = Pipeline([
    ("scaler", MinMaxScaler()),
    ("pca", PCA(n_components=4))
])

# Fit the pipeline on the data
pipeline.fit(X)

# Save the pipeline to a file
with open("pca_pipeline.pkl", "wb") as file:
    pickle.dump(pipeline, file)

# Create the clustering model
model = KMeans(n_clusters=3)

# fit the model
model.fit(pipeline.transform(X))

# Save the model to a file
with open("clustering_model_2.pkl", "wb") as file:
    pickle.dump(model, file)
```
